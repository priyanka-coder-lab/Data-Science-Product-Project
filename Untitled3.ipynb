```python
import pandas as pd
import matplotlib.pyplot as plt
import statistics
```

```python
data = {
    "City": ["Pune", "Mumbai", "Nagpur", "Nashik", "Nagar"],
    "Weather": ["Sunny", "Rainy", "Windy", "Cloudy", "Sunny"]
}

df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
weather_count = df["Weather"].value_counts()
weather_count
```

```python
most_common_weather = statistics.mode(df["Weather"])
most_common_weather
```

```python
weather_count.plot(kind='bar')
plt.xlabel("Weather Condition")
plt.ylabel("Number of Cities")
plt.title("Weather Condition Distribution")
plt.show()
```

```python
plt.bar(df["City"], df["Weather"].astype('category').cat.codes)
plt.xlabel("City")
plt.ylabel("Weather Code")
plt.title("City-wise Weather Conditions")
plt.show()
```

```python
print("Most Common Weather:", most_common_weather)
print("Weather Count:")
print(weather_count)
```
