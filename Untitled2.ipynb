```python
import pandas as pd
import matplotlib.pyplot as plt
import statistics
```

```python
data = {
    "Product": ["Mobile", "Laptop", "Headphones", "Smartwatch", "Tablet"],
    "Sales": [150, 80, 200, 120, 60]
}

df = pd.DataFrame(data)
df
```

```python
df.head()
```

```python
df.tail()
```

```python
mean_sales = df["Sales"].mean()
mean_sales
```

```python
median_sales = df["Sales"].median()
median_sales
```

```python
mode_sales = statistics.mode(df["Sales"])
mode_sales
```

```python
min_sales = df["Sales"].min()
max_sales = df["Sales"].max()

min_sales, max_sales
```

```python
best_product = df.loc[df["Sales"].idxmax()]
worst_product = df.loc[df["Sales"].idxmin()]

best_product, worst_product
```

```python
plt.bar(df["Product"], df["Sales"])
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Product Sales Analysis")
plt.show()
```

```python
plt.plot(df["Product"], df["Sales"], marker='o')
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Sales Trend")
plt.show()
```

```python
print("Average Sales:", mean_sales)
print("Highest Selling Product:", best_product["Product"])
print("Lowest Selling Product:", worst_product["Product"])
```
